```python
# recommender system
# i will try to recommend games, based on some parameters (that will be taken during the initial conversation flow)
# thhese parameters (for now) are:
# 1. Genres from the users
# 2. F2P or not
# 3. Multiplayer or not.

user_genres = ['Action', 'Free to Play', 'Strategy']
user_f2p_choice = 0
user_multiplayer_choice = 0
```

```python
similar_users = []
def find_similar_users(genres, f2p_choice, multiplayer_choice):
    if(f2p_choice == 1 & ('Free to play' not in genres)):
        genres.append('Free to play')
    if(multiplayer_choice == 1 & ('Massively Multiplayer' not in genres)):
        genres.append('Massively Multiplayer')
    for index, row in data.iterrows():
        if all(item in genres for item in row['tags']):
            similar_users.append(row)
            
    similar_users_df = pd.DataFrame(similar_users)
    return similar_users_df
```

```python
similar_users_df = find_similar_users(user_genres, user_f2p_choice, user_multiplayer_choice)
```

```python
# find out the games that were user_suggested
similar_users_recs = pd.DataFrame(similar_users_df[similar_users_df['user_suggestion'] == 1])
similar_users_recs
```

```python
# we'll simply find the top recommended games based on what games were recommended the most.
top_recs = similar_users_recs['title'].value_counts().sort_values(ascending=False)
top_recs
```

```python
similar_users_games_scores = similar_users_recs['title'].value_counts() / len(similar_users_recs)
# now lets say that we take only top 10%
# similar_users_games_scores = similar_users_games_scores[similar_users_games_scores > .10]
similar_users_games_scores
```

```python
all_users = data[(data['title'].isin(similar_users_recs['title']))]
all_users_game_scores = similar_users_recs['title'].value_counts() / len(all_users)

all_users_game_scores
```

```python
rec_percentages = pd.concat([similar_users_games_scores, all_users_game_scores],axis=1)
rec_percentages.columns = ['similar','all']

rec_percentages
```

```python
rec_percentages["score"] = rec_percentages["similar"] / rec_percentages["all"]
```

```python
rec_percentages = rec_percentages.sort_values("score", ascending=False)
```
